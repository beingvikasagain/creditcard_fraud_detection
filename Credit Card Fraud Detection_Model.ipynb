```python
data.head(5)
```

```python
data = data.drop(['Customer_ID', 'Count'], axis=1)
data.columns
```

```python
data['Default_On_Payment'].value_counts()
```

```python
#class balancing
#make it same rows for two values 0 & 1
li = list(data[data.Default_On_Payment == 0].sample(n=2000).index)
data = data.drop(data.index[li])
data['Default_On_Payment'].value_counts()
```

```python
#reset the index after drop
data.reset_index(drop=True,inplace=True)
```

```python
# finding outliers
import matplotlib.pyplot as plt
plt.scatter(data.index, data['Credit_Amount'])
```

```python
#remove 5 outliers of Credit_Amount
li = list(data['Credit_Amount'].sort_values()[-5:].index)
data['Credit_Amount'][li] = int(data.drop(li)['Credit_Amount'].mean())
plt.scatter(data.index, data['Credit_Amount'])
```

```python
#feature selection 
X = data.drop(['Default_On_Payment'], axis=1)
y = data['Default_On_Payment']
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

X = add_constant(X)
print(X.shape)
pd.Series([variance_inflation_factor(X.values, i) 
               for i in range(X.shape[1])], 
              index=X.columns).sort_values()
```

```python
#significancy (find p-value) using chi-square
import scipy.stats as stats
from scipy.stats import chi2_contingency

Significance = pd.Series()

for i in X.columns:
    dfObserved = pd.crosstab(X[i],y)
    chi2, p, dof, expected = stats.chi2_contingency(dfObserved.values)
    Significance[i] = p

Significance.sort_values()
```

```python
#cinfidence score %
import scipy.stats as stats
from scipy.stats import chi2_contingency

Significance = pd.Series()

for i in X.columns:
    dfObserved = pd.crosstab(X[i],y)
    chi2, p, dof, expected = stats.chi2_contingency(dfObserved.values)
    #print(i,p)
    Significance[i] = (1-p)*100

Significance.sort_values()
```

```python
#age and credit amount is a contineous variable . we will check anova for this
import scipy.stats as stats
stats.f_oneway(X['Credit_Amount'],y)
```

```python
import scipy.stats as stats
stats.f_oneway(X['Age'],y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=.5, random_state = 4589)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
pred
```

```python
model.score(x_test, y_test)*100
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(y_test,pred)*100
```

```python
confusion_matrix(y_test,pred)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict_proba(x_test)
```

```python
ns_probs = [0 for _ in range(len(y_test))]
lr_probs = model.predict_proba(x_test)
lr_probs = lr_probs[:, 1]
```

```python
from sklearn.metrics import roc_curve
from matplotlib import pyplot
ns_fpr, ns_tpr, _ = roc_curve(y_test, ns_probs)
lr_fpr, lr_tpr, _ = roc_curve(y_test, lr_probs)
# plot the roc curve for the model
pyplot.plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill')
pyplot.plot(lr_fpr, lr_tpr, marker='.', label='Logistic')
# axis labels
pyplot.xlabel('False Positive Rate')
pyplot.ylabel('True Positive Rate')
# show the legend
pyplot.legend()
# show the plot
pyplot.show()
```
